```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
legend_labels = ['Died', 'Survived']

ax =sns.countplot(data=df, x='Sex', hue='Survived', palette='viridis')
ax.legend(legend_labels)

plt.title('Survival by Sex')
plt.tight_layout()
plt.show()
```

```python
ax = sns.countplot(data=df, x='Pclass', hue='Survived', palette='inferno')
ax.legend(legend_labels)
ax.set_xticklabels(['1st', '2nd', '3rd'])

plt.title('Survival by Pclass')
plt.tight_layout()
plt.show()
```

```python
ax = sns.histplot(data=df, x='Age', hue='Survived', alpha=0.5, palette='plasma', multiple='dodge')
ax.legend(legend_labels)

plt.title('Age Distribution by Survival')
plt.tight_layout()
plt.show()
```

```python
ax = sns.countplot(data=df, x='Embarked', hue='Survived', palette='hls')
ax.legend(legend_labels)
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Southampton', 'Cherbourg', 'Queenstown'])
ax.set_xlabel('Port')

plt.title('Survival by Port Embarked')
plt.tight_layout()
plt.show()
```

```python
ax = sns.boxplot(data=df, x='Survived', y='Fare', palette='inferno', orient='v')
ax.legend(legend_labels)
ax.set_xticks([0, 1])
ax.set_xticklabels(['Died', 'Survived'])

plt.ylim(0, 250)
plt.title('Fare Distribution by Survival')
plt.tight_layout()
plt.show()
```

```python
ax = sns.heatmap(data=df.drop('PassengerId', axis=1).corr(numeric_only=True), cmap='coolwarm', annot=True)

plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
