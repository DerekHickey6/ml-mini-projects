```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../data/train.csv')
```

```python
legend_labels = ['Died', 'Survived']

axes =sns.countplot(data=df, x='Sex', hue='Survived', palette='viridis')
axes.legend(legend_labels)

plt.title('Survival by Sex')
plt.tight_layout()
plt.show()
```

```python
axes = sns.countplot(data=df, x='Pclass', hue='Survived', palette='inferno')
axes.legend(legend_labels)
axes.set_xticklabels(['1st', '2nd', '3rd'])

plt.title('Survival by Pclass')
plt.tight_layout()
plt.show()
```

```python
axes = sns.histplot(data=df, x='Age', hue='Survived', alpha=0.5, palette='plasma', multiple='dodge')
axes.legend(legend_labels)

plt.title('Age Distribution by Survival')
plt.tight_layout()
plt.show()
```

```python
axes = sns.countplot(data=df, x='Embarked', hue='Survived', palette='hls')
axes.legend(legend_labels)
axes.set_xticklabels(['Southampton', 'Cherbourg', 'Queenstown'])
axes.set_xlabel('Port')

plt.title('Survival by Port Embarked')
plt.tight_layout()
plt.show()
```
